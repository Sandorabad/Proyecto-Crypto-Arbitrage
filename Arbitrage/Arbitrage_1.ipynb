```python
#Setting up the requirements
apikey = '' #Correspondiente a binance
secret = '' #Correspondinte a binance
```

```python
from binance import Client, ThreadedWebsocketManager, ThreadedDepthCacheManager
import pandas as pd
import numpy as np
import networkx as nx
```

```python
client = Client(apikey, secret) #conectamos nuestra cuenta de binance
```

```python
###vamos a obtener los datos para nuestro dataframe
tickers = client.get_all_tickers()
tickers_df = pd.DataFrame(tickers)
#tickers_df.set_index('symbol', inplace=True)

#tickers que vamos a usar
names = ['XRP', 'BTC', 'ETH', 'USDT', 'USDC', 'XRP', 'ADA', 'BUSD']
```

```python
# preparamos nuestras rows
datos = []
```

```python
def tipo_cambio():
    for name in names: # por cada tickr
        for a in range(len(names)):
                for b in range(len(tickers_df.index)): #numero de rows
                    if (name + names[a] or names[a] + name) == tickers_df.iloc[b]['symbol']:
                        datos.append(tickers_df.iloc[b])
                    else:
                        continue
```

```python
def tipo_cambio_XRP():
    for i in range(tickers_df.shape[0]): #numero de rows
        if 'XRP' in tickers_df.iloc[i]['symbol']:
                XRP.append(tickers_df.iloc[i])
                
def tipo_cambio_BTC():
    for i in range(tickers_df.shape[0]): #numero de rows
        if 'BTC' in tickers_df.iloc[i]['symbol']:
                BTC.append(tickers_df.iloc[i])
                
def tipo_cambio_ETH():
    for i in range(tickers_df.shape[0]): #numero de rows
        if 'ETH' in tickers_df.iloc[i]['symbol']:
                ETH.append(tickers_df.iloc[i])
                
def tipo_cambio_USDT():
    for i in range(tickers_df.shape[0]): #numero de rows
        if 'USDT' in tickers_df.iloc[i]['symbol']:
                USDT.append(tickers_df.iloc[i])
                
def tipo_cambio_USDC():
    for i in range(tickers_df.shape[0]): #numero de rows
        if 'USDC' in tickers_df.iloc[i]['symbol']:
                USDC.append(tickers_df.iloc[i])
                
def tipo_cambio_ADA():
    for i in range(tickers_df.shape[0]): #numero de rows
        if 'ADA' in tickers_df.iloc[i]['symbol']:
                ADA.append(tickers_df.iloc[i])
                
def tipo_cambio_BNB():
    for i in range(tickers_df.shape[0]): #numero de rows
        if 'BNB' in tickers_df.iloc[i]['symbol']:
                BNB.append(tickers_df.iloc[i])
```
